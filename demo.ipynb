```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
 
data = pd.read_csv("Session2_ML_Course_Hands On.csv")
   
#print (data.head(1))
#data.shape
#drawing plot
import numpy as np
import matplotlib.pyplot as plt

x=np.array([1,2,3,4,5,6,8,10,11,15])
y=np.array([10,29,33,42,57,61,76,87,91,103])
plt.scatter(x,y)
#drawnig linear regression
from sklearn.linear_model import LinearRegression
model=LinearRegression(fit_intercept=True)
model.fit(x[:,np.newaxis],y)
xFit=np.linspace(0,30,1000)
yFit=model.predict(xFit[:,np.newaxis])
plt.scatter(x,y)
plt.plot(xFit,yFit)
```
